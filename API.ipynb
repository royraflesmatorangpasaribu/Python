```python
import json, requests
```

```python
hargaKomoditasPanganIndo = requests.get('https://jibs.my.id/api/harga_komoditas')
hargaKomoditasPanganIndo
```

```python
hargaPangan = hargaKomoditasPanganIndo.json()
hargaPangan
```

```python
def bahanPokok(namaP, komoditas):
  a = hargaPangan['national_commodity_price']
  try:
    for i in a[komoditas]:
      if i['name'] == namaP:
        print('\nBahan Pokok', komoditas, 'di Provinsi',namaP)
        print('Harga : ', i['display'])
        break
    else:
      print('\nProvinsi Tidak Ditemukan!!!')    
  except KeyError:
    print('\nBahan Pokok Tidak Ditemukan!!!')

print('Notifikasi : \nUntuk Inputan Bahan Pokok selalu di awali dengan Huruf Kapital\nDan inputan Nama Provinsi disesuaikan nama Provinsi yang ada di Data')
komoditas = input('\nMasukkan Nama Bahan Pokok\t: ')
namaP = input('Masukkan Nama Provinsi\t\t: ')
bahanPokok(namaP, komoditas)
```
